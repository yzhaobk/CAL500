```python
import pandas as pd
import numpy as np


data = np.loadtxt('cal500',delimiter = ',')
data.shape
with open('CAL500_label') as f:
    label = f.readlines()
```

```python
df = pd.DataFrame(data, columns = label)
df
```

```python
dfT = df.T
dfT
```

```python

dfT.shape
writer = pd.ExcelWriter('data_with_label.xlsx')
dfT.to_excel(writer,'Sheet1')
writer.save()
```
